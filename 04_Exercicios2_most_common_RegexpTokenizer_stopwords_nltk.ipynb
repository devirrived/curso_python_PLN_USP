```python
corpus = open("/content/drive/MyDrive/arquivos_txt/corpus_teste.txt").read()
nltk.word_tokenize(corpus)
```

```python
len(corpus)
```

```python
from nltk.tokenize import RegexpTokenizer
meutokenizer_palavras = RegexpTokenizer(r'[a-zA-Z]\w*')
palavras = meutokenizer_palavras.tokenize(corpus)
palavras
```

```python
len(palavras)
```

```python
frequencia  = nltk.FreqDist(palavras)
frequencia.most_common()
```

```python
frequencia.most_common(5)
```

```python
from nltk.tokenize import RegexpTokenizer

meutokenizer_palavras = RegexpTokenizer(r'[a-zA-Z]\w*')
tokens = meutokenizer_palavras.tokenize(corpus)

stopwords = nltk.corpus.stopwords.words('portuguese')

palavras = [token.lower() for token in tokens if token.lower() not in stopwords]

print(palavras)

frequencia = nltk.FreqDist(palavras)
frequencia.most_common()
```
